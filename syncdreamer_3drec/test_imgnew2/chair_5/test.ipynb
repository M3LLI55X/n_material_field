```python
import numpy as np
import matplotlib.pyplot as plt
   #  SDF 
sdf_field = np.load('/dtu/blackhole/11/180913/test/ViewFusion/syncdreamer_3drec/test_imgnew2/chair_5/sdf_field.npy')

   # 
print("Shape:", sdf_field.shape)
print("Data type:", sdf_field.dtype)
print("Min value:", np.min(sdf_field))
print("Max value:", np.max(sdf_field))
print("Mean value:", np.mean(sdf_field))

# 
plt.figure(figsize=(10, 6))
plt.hist(sdf_field.flatten(), bins=100)
plt.title('SDF Values Distribution')
plt.xlabel('SDF Value')
plt.ylabel('Frequency')
plt.show()

# 
center_slice = sdf_field[sdf_field.shape[0]//2, :, :]
plt.figure(figsize=(10, 10))
plt.imshow(center_slice, cmap='viridis')
plt.colorbar(label='SDF Value')
plt.title('Center Slice of SDF Field')
plt.show()

```

```python

import numpy as np
import plotly.graph_objects as go


#  3D 
fig = go.Figure(data=go.Volume(
    x=np.arange(sdf_field.shape[0]),
    y=np.arange(sdf_field.shape[1]),
    z=np.arange(sdf_field.shape[2]),
    value=sdf_field.flatten(),
    isomin=np.min(sdf_field),
    isomax=np.max(sdf_field),
    opacity=0.1,  # 
    surface_count=17,  # 
    colorscale='RdBu'  # 
))

# 
fig.update_layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    title='3D Visualization of SDF Field'
)

# 
fig.show()
```

```python
subset = sdf_field[:100, :100, :100]  #  100x100x100 

#  3D 
fig = go.Figure(data=go.Volume(
    x=np.arange(100),
    y=np.arange(100),
    z=np.arange(100),
    value=subset.flatten(),
    opacity=0.1,
    surface_count=17,
    colorscale='RdBu'
))

fig.update_layout(title='Subset of SDF Field')
fig.show()
```
